```julia
using Distributions,CSV,DataFrames,StatsPlots
```

```julia
pwd()
```

```julia
import ExponentialAnalysis as ea
```

```julia
cc3 = CSV.read("CCHANNEL_DL_3AREA_400-500.csv", DataFrame)
ccdl= cc3[:,:TTL] ;
mccgrp=  cc3[:,[:GRPT, :GRPO, :GRPN]] |> Matrix
ms = [43_000_000 , 20_000_000, 10_000_000];
```

```julia
plot(ccdl, lw=2,label="DL")
bar!(mccgrp .* 100, lw=0, alpha=0.5, label=["tokyo" "osaka" "nagoya"])
```

```julia
model=ea.expmodel(ccdl, mccgrp, ms)
```

```julia
ps,chain = ea.analysis(model)
```

```julia
chain
```

```julia
ps
```

```julia
rs_eff, lb = ea.describe_effect(ccdl, mccgrp, ms, ps)
rs_eff_mat = sum(rs_eff,dims=3) |> t-> accumulate(+, t, dims=2) |> t->reshape(t, :, 3) 
plot(rs_eff_mat , label=lb,lw=2)
plot!(ccdl, label="ccdl")
```
